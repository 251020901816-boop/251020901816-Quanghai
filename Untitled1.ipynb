```python
import math

# Nhập hệ số từ người dùng
s = input("Enter 3 real number (separated by space): ")
list_s = s.split()
a = float(list_s[0])
b = float(list_s[1])
c = float(list_s[2])

# Biện luận và giải phương trình
if a == 0:
    if b == 0:
        if c == 0:
            print("Phương trình có vô số nghiệm")
        else:
            print("Phương trình vô nghiệm")
    else:
        x = -c / b
        print(f"Phương trình bậc nhất có nghiệm duy nhất: x = {x}")
else:
    delta = b**2 - 4*a*c
    if delta > 0:
        x1 = (-b + math.sqrt(delta)) / (2*a)
        x2 = (-b - math.sqrt(delta)) / (2*a)
        print(f"Phương trình có hai nghiệm phân biệt: x1 = {x1}, x2 = {x2}")
    elif delta == 0:
        x = -b / (2*a)
        print(f"Phương trình có nghiệm kép: x1 = x2 = {x}")
    else:
        print("Phương trình vô nghiệm trong tập số thực")
```
